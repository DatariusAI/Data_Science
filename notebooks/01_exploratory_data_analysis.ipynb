# Exploratory Data Analysis: Tips Dataset

This notebook demonstrates a comprehensive EDA workflow using seaborn's built-in `tips` dataset. We cover:
1. Data overview and summary statistics
2. Distribution analysis
3. Correlation analysis
4. Categorical breakdowns
5. Multivariate exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
sns.set_palette('Set2')
%matplotlib inline
```

## 1. Data Overview

The tips dataset records restaurant tips with information about total bill, tip amount, sex of the payer, smoker status, day, meal time, and party size.

```python
tips = sns.load_dataset('tips')
print(f"Shape: {tips.shape}")
print(f"\nColumn types:")
print(tips.dtypes)
print(f"\nMissing values: {tips.isnull().sum().sum()}")
tips.head(10)
```

```python
tips.describe()
```

## 2. Distribution Analysis

Understanding the shape of numeric variables helps identify skewness, outliers, and typical ranges.

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

sns.histplot(tips['total_bill'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Total Bill Distribution')

sns.histplot(tips['tip'], bins=25, kde=True, ax=axes[1], color='coral')
axes[1].set_title('Tip Distribution')

tips['tip_pct'] = tips['tip'] / tips['total_bill'] * 100
sns.histplot(tips['tip_pct'], bins=30, kde=True, ax=axes[2], color='green')
axes[2].set_title('Tip Percentage Distribution')

plt.tight_layout()
plt.show()
```

## 3. Correlation Analysis

The heatmap reveals which numeric variables are linearly related.

```python
numeric_cols = tips.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, fmt='.3f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

## 4. Categorical Breakdowns

Box plots and violin plots show how numeric variables differ across categories.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.boxplot(data=tips, x='day', y='total_bill', ax=axes[0, 0])
axes[0, 0].set_title('Total Bill by Day')

sns.boxplot(data=tips, x='day', y='tip', hue='sex', ax=axes[0, 1])
axes[0, 1].set_title('Tips by Day and Sex')

sns.violinplot(data=tips, x='smoker', y='tip_pct', hue='time', split=True, ax=axes[1, 0])
axes[1, 0].set_title('Tip % by Smoker Status and Time')

sns.barplot(data=tips, x='day', y='tip', hue='smoker', ci='sd', ax=axes[1, 1])
axes[1, 1].set_title('Average Tip by Day and Smoker')

plt.tight_layout()
plt.show()
```

## 5. Multivariate Exploration

Pairplots and scatter plots with hue reveal complex relationships between multiple variables.

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker', size='size',
                sizes=(30, 200), alpha=0.7)
plt.title('Total Bill vs Tip (by Time, Smoker, and Party Size)')
plt.tight_layout()
plt.show()
```

```python
g = sns.pairplot(tips[['total_bill', 'tip', 'size', 'tip_pct', 'time']], hue='time',
                 plot_kws={'alpha': 0.5, 's': 30})
g.figure.suptitle('Pairplot by Meal Time', y=1.02)
plt.show()
```

## Key Findings

- Total bill and tip are strongly positively correlated (~0.68)
- Tip percentage averages around 15-17%, with some generous outliers
- Weekend dinners tend to have higher total bills
- Party size has a moderate positive relationship with total bill
- Smoker status does not appear to have a major effect on tipping behavior
